### Question 5

Now, build your scrambler. Assume the user will enter a word, or a name, or even a phrase. Scramble it and return the input scrambled after the user type it.

````python
# Your answer here
import random as rd
def scrambler():
    entry = input("enter your word, name or phrase")
    entry_list = list(entry)
    entry_list_rd = rd.sample(entry_list, len(entry_list))
    print(entry_list_rd)
    return entry_list_rd
scrambler()
````
